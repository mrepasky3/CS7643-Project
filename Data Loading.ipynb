```python
import torch
import torchvision
import torchvision.datasets as datasets
import matplotlib.pyplot as plt
import dataset_prep
```

# Toy

```python
set_loader = dataset_prep.SetsToy(max_size=4,fixed_size=True)
train_d, train_l, test_d, test_l = set_loader.sum_task()
```

# MNIST

```python
set_loader = dataset_prep.SetsMNIST()
train_d, train_l, test_d, test_l = set_loader.sum_task()
```

```python
num = torch.randint(train_l.shape[0],(1,))
print(train_l[num])
for i in range(len(train_d[num])):
    fig = plt.figure(figsize=(1,1))
    plt.imshow(train_d[num][i])
    plt.show()
```
